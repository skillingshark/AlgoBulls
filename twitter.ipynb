```python
from selenium import webdriver 
from selenium.webdriver.common.by import By
import re
import time 
import pandas as pd
import random
```

```python

driver = webdriver.Chrome()

url = 'https://twitter.com/search?q=%23coding&src=typed_query'
driver.get(url)
```

```python
# Regular expressions to extract handle, content, and likes using the observed patterns

handle_pattern = r'@(\w+)' # Extract handle by matching @
likes_pattern = r'(\d+)\n\d+$'  # Match the second last number as it is the number of likes 
content_pattern = r'(?<=·\n).*?(?=\n\d+\n\d+$)'  # Match the content after "·\n" and before the last two numbers

twitter_handles = []
tweet_content = []
likes_list = []

# Extracting 200 entries as some of them would be repeated due overlap of scrolls
while(len(twitter_handles)<200):

    # Scroll down by 2 times viewheight to get new content and minimize overlap
    driver.execute_script("window.scrollBy(0, 2 * window.innerHeight)")
    time.sleep(2)
    
    # list of web elements with contain all text data corresponding to a tweet 
    all_content = driver.find_elements(By.CSS_SELECTOR,  ".css-175oi2r.r-1iusvr4.r-16y2uox.r-1777fci.r-kzbkwu")
    for i in all_content:
        # Extracting twitter_handle, tweet_content, and likes
        try:
            handle = re.search(handle_pattern, i.text).group(0)
            content = re.search(content_pattern, i.text, re.DOTALL).group(0)
            likes = re.search(likes_pattern, i.text, re.MULTILINE).group(1)
            
            twitter_handles.append(handle)
            tweet_content.append(content)
            likes_list.append(likes)
            print(len(twitter_handles) , handle) # Printing to check progress
        except:
            pass
    
# print(twitter_handles, tweet_content , likes)
```

```python
print(len(set(tweet_content)))
```

```python
# Storing the data in a pandas dataframe

df = pd.DataFrame({'Twitter Handle' : twitter_handles , 'Likes' : likes_list , 'Tweet Content' : tweet_content})
df.to_csv('twitter_extract')
df
```
